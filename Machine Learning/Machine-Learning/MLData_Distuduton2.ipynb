```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cwt = np.random.normal(20.0,2.0,1000)
```

```python
print(cwt)
```

```python
x = cwt.mean()
```

```python
print(x)
```

```python
y = np.std(cwt)
```

```python
print(y)
```

```python
mi = np.min(cwt)
```

```python
print(mi)
```

```python
ma = np.max(cwt)
```

```python
print(ma)
```

```python
plt.hist(cwt, 50)
plt.title('Data')
plt.xlabel('STD')
plt.ylabel('MIN')
```
