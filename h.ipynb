```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
system = pd.read_csv('ghana_school_fees_data.csv')
system
```

```python
system.info()
system.dropna(inplace = True)
system
```

```python
system.drop_duplicates(inplace = True)
system
```

```python
plt.hist(system['Fee'], bins = 10,color = 'black')
x = system['Students']
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, system['Fee'])

def fun(x):
    return slope * x + intercept

model = list(map(fun,x))
```

```python
plt.scatter(x,system['Fee'],marker = 'x',color = 'black')
plt.plot(x,model,color = 'green')
```

```python
hunstudents = fun(100)
thunstudents = fun(200)
thrhundredstudents = fun(300)
fhundredstudents = fun(100)
print(f'The revenue for a 100 students is {hunstudents}')
print(f'The revenue for a 200 students is {thunstudents}')
print(f'The revenue for a 300 students is {thrhundredstudents}')
print(f'The revenue for a ? students is {fhundredstudents}')
```
