```python
data_loader['Total'] = data_loader[['Math','English','Science','ICT','Social_Studies']].sum(axis = 1)
```

```python
data_loader.head()
```

```python
data_loader['Average'] = data_loader[['Total']]/5
```

```python
data_loader.head()
```

```python
data_loader
```

```python
data_loader[["Math"]]
```

```python
subjects = ['Math','English','Science','ICT','Social_Studies']
```

```python
for subject in subjects:
    avg = data_loader[subject].mean()
    print(subject, avg)
    
lower_avg = min(subjects, key = lambda sugject: avg)
print("Lowest",lower_avg, avg)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(data_loader['Participation'], data_loader['Total'], color = "green")
plt.xlabel("Participation")
plt.ylabel("Total")
plt.grid(True)
plt.show()
```

```python
avg_scores = [data_loader[subject].mean() for subject in subjects]

plt.bar(subjects, avg_scores, color='red')
plt.title("Average Score per Subject")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
