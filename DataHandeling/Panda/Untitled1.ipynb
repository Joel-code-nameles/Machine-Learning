```python
import pandas
import matplotlib.pyplot as plt
```

```python
system = pandas.read_csv('health_data.csv')
```

```python
system
```

```python
avg_age = system['Age'].mean()
avg_weight = system['Weight_kg'].mean()
avg_systolic = system['Systolic_BP'].mean()
avg_Cholestrol = system['Cholesterol_mg_dL'].mean()
```

```python
system['Age'].fillna(avg_age, inplace=True)
system['Weight_kg'].fillna(avg_weight, inplace=True)
system['Systolic_BP'].fillna(avg_systolic, inplace=True)
system['Cholesterol_mg_dL'].fillna(avg_Cholestrol, inplace=True)
system
```

```python
system.plot()
```

```python
system.plot(kind = 'hist')
```

```python
plt.figure(figsize=(8,5))

plt.scatter(system['Age'], system['Cholesterol_mg_dL'], color = "black")
plt.xlabel("Attendance")
plt.ylabel("Total Grade")
plt.title("Does Attendance affect Grades")
plt.grid(True)
```
