```python
import pandas
import matplotlib.pyplot as plt

dataframe = pandas.read_csv('student_scores.csv')
dataframe.head()
```

```python
dataframe.tail()
```

```python
dataframe['Total'] = dataframe[['Math','English','Science',"ICT","Social_Studies"]].sum(axis = 1)
```

```python
dataframe.head()
```

```python
dataframe['Average'] = dataframe[['Total']]/5
```

```python
dataframe.head()
```

```python
dataframe
```

```python
subjects = ['Math','English','Science','ICT','Social_Studies']
```

```python
for subject in subjects:
    avg = dataframe[subject].mean()
    print(f"Subject: {subject}: Averge: {avg:.2f}")
```

```python
lowest_avg = min(subjects, key = lambda subjects: dataframe[subject].mean())
print(lowest_avg, avg)
```

```python
dataframe.sort_values(by = "Total" , ascending = False).head(3)
```

```python
dataframe[dataframe['Average'] < 70].sort_values(by = "Total", ascending = False)
```
