```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
system = pd.read_csv('assignment_student_data.csv')
system
```

```python
system['Total'] = system[['Math','English','Science','ICT','Social_Studies']].sum(axis = 1)
system
```

```python
system['Average'] = system[['Total']]/5
system
```

```python
highest_score = system.sort_values(by = "Total", ascending = False).head(1)
highest_score
```

```python
lowest_score = system[system['Average'] < 70].sort_values(by = "Total", ascending = True).head(1)
lowest_score
```

```python
plt.figure(figsize=(8,5))

plt.scatter(system['Attendance'], system['Total'], color = "black")
plt.xlabel("Attendance")
plt.ylabel("Total Grade")
plt.title("Does Attendance affect Grades")
plt.grid(True)
```

```python
#Yes, Indubently the graph shows a strong positive relationship between attendance and grades.
#The more a student attends school, the higher their total grade tends to be,
#likely because regular attendance boosts learning and retention.
```
