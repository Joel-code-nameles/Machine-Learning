```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
liger = pd.read_csv('housing_prices_linear_regression.csv')
```

```python
liger
```

```python
plt.scatter(liger['Size (sqft)'], liger['Price ($1000s)'],marker = 'x',color = 'blue')
```

```python
slope, intercept, r_value, p_value, std_error = stats.linregress(liger['Size (sqft)'], liger['Price ($1000s)'])
x = liger['Size (sqft)']
```

```python
def fun(x):
    return slope * x + intercept
```

```python
model = list(map(fun, x))
prediction_of_10000 = fun(10000)
prediction_of_50000 = fun(50000)
print(f'Prediction of 10000 sqft house: {prediction_of_10000}')
print(f'Prediction of 50000 sqft house: {prediction_of_50000}')
```

```python
plt.scatter(liger['Size (sqft)'], liger['Price ($1000s)'],marker = 'x',color = 'blue')
plt.title('Housing price based on size')
plt.xlabel('Size')
plt.ylabel('Price')
plt.plot(x, model)
```
